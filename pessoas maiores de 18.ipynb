# Importação das bibliotecas

```python
import pandas as pd
from datetime import datetime
```

# Importando e configurando o DataFrame do arquivo `.csv`

```python
df = pd.read_csv('/content/people.csv')
df['birthday'] = pd.to_datetime(df['birthday']) # configura a coluna birthday como uma data
df
```

# Função para calcular a idade

```python
def calc_idade(data_nasc):
  hoje = datetime.today()

  return hoje.year - data_nasc.year - ((hoje.month, hoje.day) < (data_nasc.month, data_nasc.day))
```

# Adicionando a coluna `age` no DataFrame

```python
df['age'] = df['birthday'].apply(calc_idade)
df['age'] = pd.to_numeric(df['age'])
df
```

# Conferindo as informações do DataFrame

```python
df.info()
```

```python
try:
  with open('/content/pessoas_maiores_18.txt', 'w') as arq_texto:
    df_ordenado = df[['name','age']].sort_values(by='name')
    df_ordenado = df_ordenado[df_ordenado['age'] >= 18]

    arq_texto.write(f'Existem {len(df_ordenado)} pessoas maiores de 18 anos.\n')

    for i in df_ordenado.values:
      arq_texto.write(f'\n{i[0]} tem {i[1]} anos.')

    print('Arquivo salvo com sucesso!\n')
    display(df_ordenado)

except Exception as e:
  print(f'Erro desconhecido ao salvar o arquivo. Por favor, tente novamente mais tarde.')

finally:
  arq_texto.close()
```
